1. Basic Line Plot
Plot a simple line graph of ùë¶ = ùë•^2 for ùë• values ranging from -10 to 10.
Label the x-axis as "X Values" and the y-axis as "Squared Values."
Add a title to the graph.

```python
import matplotlib.pyplot as plt
import numpy as np
x = np.linspace(-10, 10, 100) #Generates 100 random values from -10 to 10
y = x**2
plt.plot(x, y)
plt.xlabel('X Values')
plt.ylabel('Squared Values')
plt.title('Basic Line Plot for y=x^2')  #Generates heading for the Graph
plt.show() #For displaying
```

4. Histogram
Simulate 100 random values from a normal distribution using np.random.randn() and plot a histogram.
Customize the number of bins and add a grid to the plot.

```python
import  matplotlib.pyplot as plt
import numpy as np
data = np.random.randn(100)
plt.hist(data, bins=10, edgecolor='black')
plt.xlabel('Value')
plt.ylabel('Frequency')
plt.title('Histogram of Random Values')
plt.grid(True)  #Adds rows and columns as grid on the whole graph
plt.show()
```

5. Pie Chart
Visualize the percentage distribution of five categories ['A', 'B', 'C', 'D', 'E'] with values [20, 30, 25, 15, 10].
Display the percentage values on the chart and use a color palette for differentiation

```python
import matplotlib.pyplot as plt
categories = ['A', 'B', 'C', 'D', 'E']
values = [20, 30, 25, 15, 10]
colors = ['blue', 'green', 'red', 'orange', 'purple']
plt.pie(values, labels=categories, colors=colors, autopct='%1.1f%%')
plt.title('Percentage Distribution of Categories')
plt.show()
```

8.Comparison of Matplotlib and Seaborn
Generate a pair of plots (bar charts or line plots) using the same data in both Matplotlib and Seaborn.
Highlight the differences in the code, syntax, and aesthetics of the two libraries

```python
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np

# Data for the plots
x = np.linspace(0, 10, 100)
y1 = np.sin(x)

# Line plot using Matplotlib
plt.figure(figsize=(4,4))
plt.plot(x, y1)
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.title('Line Plot using Matplotlib')
plt.show()

# Line plot using Seaborn
plt.figure(figsize=(4,4))
sns.lineplot(x=x, y=y1, color='red')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.title('Line Plot using Seaborn')
plt.show()
```

10. Save and Style Plots
Plot any graph of your choice and save it in both .png and .pdf formats.
Apply at least one built-in style (e.g., 'ggplot' or 'seaborn-darkgrid') and explain how the style impacts the visualization.

```python
import matplotlib.pyplot as plt
import numpy as np
x = np.linspace(0, 10, 100)
y1 = np.sin(x)
y2 = np.cos(x)
plt.plot(x, y1, label='sin(x)',color='red')
plt.plot(x, y2, label='cos(x)')
plt.xlabel('x')
plt.ylabel('y')
plt.title('Sine and Cosine Functions')
plt.legend()
plt.savefig('plot.png')
plt.savefig('plot.pdf')
plt.show()
```
