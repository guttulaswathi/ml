```python
sns.pairplot(data_cleaned, diag_kind="hist", corner=True)
plt.suptitle("Pairplot of Advertising Data", y=1.02)
plt.show()
```

# Pairplot to visualize relationships
TV and Sales: A positive correlation is visible; higher TV advertisement budgets are associated with higher sales.

Radio and Sales: A similar positive trend is observed, suggesting an increase in sales with higher radio advertisement spending.

Newspaper and Sales: A weaker correlation compared to TV and Radio.

TV, Radio, and Newspaper: Minimal interaction is seen between these variables.


```python
plt.figure(figsize=(8, 7))
correlation = data_cleaned.corr()
sns.heatmap(correlation, annot=True, cmap="coolwarm", fmt=".2f")
plt.title("Correlation Heatmap")

```

# Correlation Heatmap:
TV and Sales show the strongest correlation (0.780).

Radio also correlates moderately with Sales (0.58).

Newspaper has a weak correlation with Sales (0.23).
